### Building Recommendation System

```python
# Identify our columns
movies.columns
```

```python
# Definying x and y variable

x = movies[['IMDB Rating', 'Biography', 'Drama',
       'Thriller', 'Comedy', 'Crime', 'Mystery', 'History']]
y = movies['Movie Name']
```

```python
# Implement nearestneighbor
# Source: https://scikit-learn.org/stable/modules/generated/sklearn.neighbors.NearestNeighbors.html
# Source: https://scikit-learn.org/stable/modules/generated/sklearn.neighbors.KNeighborsClassifier.html

nbrs = NearestNeighbors(n_neighbors=5).fit(x)
distances, indices = nbrs.kneighbors(x)
print(distances)
```

Genre Information: "The Post"

IMDB Rating = 7.2, Biography = Yes, Drama = Yes, Thriller = No, Comedy = No, Crime = No, Mystery = No, History = Yes

```python
#define variable for features of "The Post"
the_post = [7.2,1,1,0,0,0,0,1]

distances, indices = nbrs.kneighbors([the_post])
print(indices)
```

```python
# Find recommended movies related to "The Post"
for title in indices:
  print("Movies you might enjoy:")
  print(movies.iloc[title]['Movie Name'])
```

The movies above are recommendations based on watching "The Post". Similar to what Netflix uses after you have finished a show or movie, the movies above will be recommended based on your user patterns.
